### CREDIT CARD FRAUD DETECTION..

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("creditcard.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#number of legit and fraudlent transactions
df['Class'].value_counts()
```

```python
#here 0 is for genuine transaction and 1 is for fraud transactions
```

```python
#since we can see that this data is highly unbalanced  we need to perform unders sampling
#we have to firstly separate this for analysis
```

```python
leg=df[df.Class==0]
fraud=df[df.Class==1]
```

```python
leg.shape
```

```python
fraud.shape
```

```python
#comparing values for both the transaction
df.groupby('Class').mean()
```

```python
#here we can see that there is signifcant difference between the mean values
```

```python
#let's do under sampling for the normal transaction to make it similar in size with fraudlent 
```

```python
normal=leg.sample(n=492)
```

```python
#now concatenatting the dataframe
new_df=pd.concat([normal,fraud],axis=0)
```

```python
new_df.head()
```

```python
new_df.groupby('Class').mean()
```

```python
#splitting data into features and target
X=new_df.drop(columns='Class',axis=0)
Y=new_df['Class']
```

```python
X
```

```python
Y
```

```python
#splitting data into training and testing part
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
#ensures that the class distribution in the original dataset is preserved in the training 
#and testing subsets 
#randomstate ensures a element of randomness
```

```python
model=LogisticRegression(max_iter=1000)
model.fit(X_train,Y_train)
```

```python
#model evaluation
X_train_pred=model.predict(X_train)
acc=accuracy_score(X_train_pred,Y_train)
```

```python
acc*100  #accuracy on training data
```

```python
Y_test
```

```python
x_test_pred=model.predict(X_test)
ac2=accuracy_score(x_test_pred,Y_test)
ac2*100 #accuracy on testing data
```
